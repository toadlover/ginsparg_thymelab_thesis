```python
#now get jaccard similarity
import pandas as pd
from sklearn.metrics import jaccard_score
from itertools import combinations
import numpy as np
```

```python
#pip install scikit-learn
```

```python
df = pd.read_csv("../../10_drugs_being_tested_in_zebrafish/system_energies_binary_sorted_renamed_duplicates_removed.csv", index_col=0)
df
```

```python
# Prepare an empty similarity matrix
systems = df.index
similarity_matrix = pd.DataFrame(index=systems, columns=systems, dtype=float)
```

```python
# Calculate Jaccard similarity for each system pair
for i in systems:
    for j in systems:
        similarity = jaccard_score(df.loc[i], df.loc[j])
        similarity_matrix.loc[i, j] = similarity
```

```python
# Optional: Save to CSV or visualize
similarity_matrix.to_csv("fish_jaccard_similarity_matrix.csv")
print(similarity_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
ax = sns.heatmap(similarity_matrix.astype(float), cmap="viridis", annot=False, cbar_kws={"label": "Jaccard Similarity"})
plt.title("HCRTR2 Drug Binding Mode Similarity")
plt.tight_layout()

# Example: change tick label colors
# Set row (y-axis) label color
for tick_label in ax.get_yticklabels():
    if tick_label.get_text() == '7SR8_TAK-925':
        tick_label.set_color('green')
    if tick_label.get_text() == '4S0V_suvorexant':
        tick_label.set_color('magenta')
    if tick_label.get_text() == '5WQC_EMPA':
        tick_label.set_color('magenta')
    if tick_label.get_text() == '6TPN_HTL6641':
        tick_label.set_color('magenta')
    if tick_label.get_text() == '7XRR_lemborexant':
        tick_label.set_color('magenta')
    if tick_label.get_text() == '7L1V_agonist':
        tick_label.set_color('green')

for tick_label in ax.get_xticklabels():
    if tick_label.get_text() == '7SR8_TAK-925':
        tick_label.set_color('green')
    if tick_label.get_text() == '4S0V_suvorexant':
        tick_label.set_color('magenta')
    if tick_label.get_text() == '5WQC_EMPA':
        tick_label.set_color('magenta')
    if tick_label.get_text() == '6TPN_HTL6641':
        tick_label.set_color('magenta')
    if tick_label.get_text() == '7XRR_lemborexant':
        tick_label.set_color('magenta')
    if tick_label.get_text() == '7L1V_agonist':
        tick_label.set_color('green')


plt.show()
```
