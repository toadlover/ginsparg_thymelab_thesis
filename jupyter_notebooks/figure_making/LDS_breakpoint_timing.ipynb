```python
#initial import of relevant packages for graphing
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import FuncFormatter
#https://stackoverflow.com/questions/32085531/importing-a-csv-file-into-python-to-make-scatterplots-and-histograms
```

```python
#Data from 100 ligand run on 7l1u
"""
Clash time per placement 0.00042609
Total space fill time = 2.21764e+06
Space fill time per placement = 7.5203
Total initial scoring time = 607881
Init scoring time per placement = 3.27635
Total initial ddg time = 94881.5
Init ddg time per placement = 52.5368
Total hrd time = 1.1819e+06
Hrd time per placement = 654.431
Total post ddg time = 91705.8
Post ddg time per placement = 50.7784
Total post scoring time = 32593.6
Post scoring time per placement = 18.0474
Total motif time = 113550
Motif time per placement = 351.548

[Clash Check,3088.89]
[Space Fill,2.21764e+06]
[Initial Score Function,607881]
[Initial ddG,94881.5]
[HighResDock,1.1819e+06]
[Post ddG,91705.8]
[Post Score,32593.6]
[Placement Motifs,113550]

[Clash time per placement,0.00042609]
[Space fill time per placement,7.5203]
[Init scoring time per placement,3.27635]
[Init ddg time per placement,52.5368]
[Hrd time per placement,654.431]
[Post ddg time per placement,50.7784]
[Post scoring time per placement,18.0474]
[Motif time per placement,351.548]
"""
```

```python
#data in format to make total time spent graph
#times are in milliseconds initially
total_data = [["Clash Check",3088.89/1000],["Space Fill",2.21764e+06/1000],["Initial Score",607881/1000],["Initial ddG",94881.5/1000],["HighResDock",1.1819e+06/1000],["Post ddG",91705.8/1000],["Post Score",32593.6/1000],["Placement Motifs",113550/1000]]

#convert to dataframe
df = pd.DataFrame(total_data, columns=['X', 'Y'])
```

```python
# Create a bar plot
plt.bar(df['X'], df['Y'])
plt.xlabel('Ligand Discovery Search Step')
plt.ylabel('Time (log10 s)')
plt.title('Total Time Taken on Steps of LigandDiscoverySearch',y=1.05)
plt.suptitle("Search on 100 ligands", fontsize=10,y=.92)


# Set y-axis to display in scientific notation with individual tick values
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
plt.gca().yaxis.get_major_formatter().set_scientific(False)
plt.gca().yaxis.get_major_formatter().set_powerlimits((0,0))

# Set y-axis to display numbers with commas
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))

#add note for the text
#plt.text(3, 1, "Conformers: " + str(total_ligands), fontsize=12, color='black')
#plt.annotate("Total Unique Conformers = " + str(total_ligands_str), xy=(0.245, -0.18), xycoords='axes fraction', fontsize=9, color='black')

plt.yscale("log")

# Rotate the x-axis labels
plt.xticks(rotation=90)

plt.show()
```
