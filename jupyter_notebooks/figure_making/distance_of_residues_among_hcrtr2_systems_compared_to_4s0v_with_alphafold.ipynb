```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#https://stackoverflow.com/questions/32085531/importing-a-csv-file-into-python-to-make-scatterplots-and-histograms
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
df  = pd.read_csv("~/ginsparg_thymelab_thesis/aligned_hcrtr2_pdbs_for_analysis/4s0v_pocket_residue_distances_with_af.csv", index_col=0)
# Drop columns with all NaN values
df_cleaned = df.dropna(axis=1, how='all')
df = df_cleaned
```

```python
#this crashes with how big the dataframe is with the open options
df
```

```python
#assign agonist and antagonist labels
antagonists = ["5wqc", "5ws3", "6tpg", "6tpj", "6tpn", "7xrr"]
agonists = ["7l1u", "7l1v", "7sqo", "7sr8"]
```

```python
# Add a "Group" column
df["Group"] = ["active" if row in antagonists else "inactive" for row in df.index]
```

```python
# Sort by group
df_sorted = df.sort_values(by="Group")
df_sorted
```

```python
# Create a heatmap
plt.figure(figsize=(20, 8))
ax = sns.heatmap(
    df_sorted.drop(columns=["Group"]),  # Exclude the "Group" column from the heatmap
    annot=False,
    cmap="viridis",
    fmt=".1f",
    annot_kws={"size": 8},
    cbar_kws={"shrink": 0.8,"orientation": "vertical", "location": "left"},
)

# Add horizontal lines to separate groups
group_counts = df_sorted["Group"].value_counts()
agonist_count = group_counts["active"]
antagonist_count = group_counts["inactive"]
ax.hlines(y=agonist_count, xmin=0, xmax=len(df_sorted.columns), color="black", linewidth=1.5)

ax.hlines(y=agonist_count + antagonist_count - 1, xmin=0, xmax=len(df_sorted.columns), color="black", linewidth=1.5)

# Add group labels
plt.text(len(df_sorted.columns) + 0.1, (agonist_count / 2) + 0.5, "Inactive", 
         fontsize=12, color="black", va="bottom", rotation=90)
plt.text(len(df_sorted.columns) + 0.1, (antagonist_count + (len(df_sorted) - antagonist_count) / 2) + 0.3, "Active", 
         fontsize=12, color="black", va="bottom", rotation=90)
plt.text(len(df_sorted.columns) + 0.1, (antagonist_count + (len(df_sorted) - antagonist_count) / 2) + 0.3 + 2.75, "Alphafold", 
         fontsize=12, color="black", va="bottom", rotation=90)

#add heatmap key label
plt.text(-6.3, 8.1, "RMSD of System Residue Against 4S0V Residue ()", 
         fontsize=12, color="black", va="bottom", rotation=90)

# Add labels and title
#plt.title("HCRTR2 Pocket Residue RMSD Against 4S0V", fontsize=16)
#plt.xlabel("Residue Indices", fontsize=12)
#plt.ylabel("Systems", fontsize=12)

plt.xticks(rotation=50, ha="right", fontsize='14')

plt.yticks(fontsize=14)

# Show the heatmap
plt.tight_layout()

plt.savefig("supplemental_figure_4s0v_residue_comparison_figure_af.svg", format="svg")

plt.show()
```
