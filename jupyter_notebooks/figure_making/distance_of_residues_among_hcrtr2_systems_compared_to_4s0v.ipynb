```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#https://stackoverflow.com/questions/32085531/importing-a-csv-file-into-python-to-make-scatterplots-and-histograms
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
df  = pd.read_csv("/home/abgvg9/ginsparg_thymelab_thesis/aligned_hcrtr2_pdbs_for_analysis/4s0v_pocket_residue_distances.csv", index_col=0)
# Drop columns with all NaN values
df_cleaned = df.dropna(axis=1, how='all')
df = df_cleaned
```

```python
#this crashes with how big the dataframe is with the open options
df
```

```python
#assign agonist and antagonist labels
antagonists = ["5wqc", "5ws3", "6tpg", "6tpj", "6tpn", "7xrr"]
agonists = ["7l1u", "7l1v", "7sqo", "7sr8"]
```

```python
# Add a "Group" column
df["Group"] = ["antagonist" if row in antagonists else "agonist" for row in df.index]
```

```python
# Sort by group
df_sorted = df.sort_values(by="Group")
```

```python
# Create a heatmap
plt.figure(figsize=(16, 8))
ax = sns.heatmap(
    df_sorted.drop(columns=["Group"]),  # Exclude the "Group" column from the heatmap
    annot=True,
    cmap="RdYlGn_r",
    fmt=".1f",
    annot_kws={"size": 8},
    cbar_kws={"shrink": 0.8,"orientation": "vertical", "location": "left"},
)

# Add horizontal lines to separate groups
group_counts = df_sorted["Group"].value_counts()
agonist_count = group_counts["agonist"]
ax.hlines(y=agonist_count, xmin=0, xmax=len(df_sorted.columns), color="black", linewidth=1.5)

# Add group labels
plt.text(len(df_sorted.columns) + 0.5, antagonist_count / 2, "Agonists", 
         fontsize=12, color="black", va="bottom", rotation=90)
plt.text(len(df_sorted.columns) + 0.5, antagonist_count + (len(df_sorted) - antagonist_count) / 2, "Antagonists", 
         fontsize=12, color="black", va="bottom", rotation=90)

# Add labels and title
plt.title("HCRTR2 Pocket Residue RMSD against 4S0V", fontsize=16)
plt.xlabel("Residue Indices", fontsize=12)
plt.ylabel("Systems", fontsize=12)

# Show the heatmap
plt.tight_layout()
plt.show()
```
