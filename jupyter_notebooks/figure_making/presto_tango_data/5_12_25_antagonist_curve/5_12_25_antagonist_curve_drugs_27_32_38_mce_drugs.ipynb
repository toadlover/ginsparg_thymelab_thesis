```python
#import packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D 
```

```python
# Step 1: Load data
drug1_data = pd.read_csv("5_12_25_d27_data.csv")
drug2_data = pd.read_csv("5_12_25_d32_data.csv")
drug3_data = pd.read_csv("5_12_25_d38_data.csv")
drug4_data = pd.read_csv("5_12_25_empa_data.csv")
drug5_data = pd.read_csv("5_12_25_MK-1064_data.csv")
drug6_data = pd.read_csv("5_12_25_o2ra_data.csv")
drug7_data = pd.read_csv("5_12_25_SB674042_data.csv")
drug8_data = pd.read_csv("5_12_25_TCS-1102_data.csv")
drug9_data = pd.read_csv("5_12_25_TCS-OX2-29_data.csv")
```

```python
#get the OxA positive control average and DMSO negative control average
#OxA average is used to normalize drugs against each other from other experiments
#DMSO average is used to show baseline

oxa_average = 1570220
dmso_average = 35130

drug1_data = drug1_data / oxa_average
drug2_data = drug2_data / oxa_average
drug3_data = drug3_data / oxa_average
drug4_data = drug4_data / oxa_average
drug5_data = drug5_data / oxa_average
drug6_data = drug6_data / oxa_average
drug7_data = drug7_data / oxa_average
drug8_data = drug8_data / oxa_average
drug9_data = drug9_data / oxa_average
```

```python
# Extract concentrations and replicates
concentrations = drug1_data.iloc[:, 0] * oxa_average # First column is the concentration, undo the divide by the oxa average
drug1_replicates = drug1_data.iloc[:, 1:]  # Columns 2-5 are replicates
drug2_replicates = drug2_data.iloc[:, 1:]  # Columns 2-5 are replicates
drug3_replicates = drug3_data.iloc[:, 1:]
drug4_replicates = drug4_data.iloc[:, 1:]
drug5_replicates = drug5_data.iloc[:, 1:]
drug6_replicates = drug6_data.iloc[:, 1:]
drug7_replicates = drug7_data.iloc[:, 1:]
drug8_replicates = drug8_data.iloc[:, 1:]
drug9_replicates = drug9_data.iloc[:, 1:]
```

```python
# Step 2: Calculate mean and standard error for each concentration
drug1_means = drug1_replicates.mean(axis=1)
drug1_sems = drug1_replicates.sem(axis=1)  # SEM = std / sqrt(n)
drug2_means = drug2_replicates.mean(axis=1)
drug2_sems = drug2_replicates.sem(axis=1)
drug3_means = drug3_replicates.mean(axis=1)
drug3_sems = drug3_replicates.sem(axis=1)
drug4_means = drug4_replicates.mean(axis=1)
drug4_sems = drug4_replicates.sem(axis=1)
drug5_means = drug5_replicates.mean(axis=1)
drug5_sems = drug5_replicates.sem(axis=1)
drug6_means = drug6_replicates.mean(axis=1)
drug6_sems = drug6_replicates.sem(axis=1)
drug7_means = drug7_replicates.mean(axis=1)
drug7_sems = drug7_replicates.sem(axis=1)
drug8_means = drug8_replicates.mean(axis=1)
drug8_sems = drug8_replicates.sem(axis=1)
drug9_means = drug9_replicates.mean(axis=1)
drug9_sems = drug9_replicates.sem(axis=1)
```

```python
#Plot the data
plt.figure(figsize=(10, 6))

# Plot drug 1
plt.errorbar(
    concentrations, drug1_means, yerr=drug1_sems, fmt="s-", label="d27",
    capsize=5, color="red"
)

# Plot drug 2
plt.errorbar(
    concentrations, drug2_means, yerr=drug2_sems, fmt="s-", label="d32",
    capsize=5, color="orange"
)

plt.errorbar(
    concentrations, drug3_means, yerr=drug3_sems, fmt="s-", label="d38",
    capsize=5, color="yellow"
)

plt.errorbar(
    concentrations, drug4_means, yerr=drug4_sems, fmt="s-", label="EMPA",
    capsize=5, color="green"
)

plt.errorbar(
    concentrations, drug5_means, yerr=drug5_sems, fmt="s-", label="SB-674042 ",
    capsize=5, color="blue"
)

plt.errorbar(
    concentrations, drug6_means, yerr=drug6_sems, fmt="s-", label="TCS-1102",
    capsize=5, color="purple"
)

plt.errorbar(
    concentrations, drug7_means, yerr=drug7_sems, fmt="s-", label="TCS-OX2-29",
    capsize=5, color="pink"
)

plt.errorbar(
    concentrations, drug8_means, yerr=drug8_sems, fmt="s-", label="MK-1064",
    capsize=5, color="black"
)

plt.errorbar(
    concentrations, drug9_means, yerr=drug9_sems, fmt="s-", label="Orexin 2 Receptor Antagonist",
    capsize=5, color="grey"
)

# Customize the plot
plt.title("Receptor Activity at Different Drug Concentrations", fontsize=16)
plt.xlabel("Concentration", fontsize=14)
plt.ylabel("Receptor Activity", fontsize=14)
plt.xscale("log")  # Log scale if concentrations span several orders of magnitude

plt.grid(alpha=0.5)
plt.tight_layout()

plt.axhline(y=oxa_average/oxa_average, color="green", linestyle="--", label="OxA 1e-7M")
plt.axhline(y=dmso_average/oxa_average, color="black", linestyle="--", label="DMSO")

# Create custom legend entries for the horizontal lines
#line1 = Line2D([0], [0], color="green", linestyle="--", label="OxA 1e-7M")
#line2 = Line2D([0], [0], color="red", linestyle="--", label="DMSO")

#plt.legend(handles=[line1, line2] + plt.gca().get_lines(), fontsize=12)
plt.legend(fontsize=12)

# Show the plot
plt.show()
```
