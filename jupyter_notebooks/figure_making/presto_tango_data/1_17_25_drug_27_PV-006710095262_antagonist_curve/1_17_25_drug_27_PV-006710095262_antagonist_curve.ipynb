```python
#initial import of relevant packages for graphing
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import FuncFormatter
#https://stackoverflow.com/questions/32085531/importing-a-csv-file-into-python-to-make-scatterplots-and-histograms
```

```python
#read in the data
df = pd.read_csv("1_17_25_drug27_antagonist_curve_data_for_plot.csv", index_col=0)  # Set the first column as row index
```

```python
df
```

```python
#get mean and std dev
row_means = df.mean(axis=1)  # Mean of each row
row_stds = df.std(axis=1)    # Standard deviation of each row
```

```python
#make plot that include 1e-6 replicate 1 
#Well may have been compromised, as well volume looked greater than other wells, and point may be worth throwing out

plt.figure(figsize=(10, 6))
plt.errorbar(row_means.index, row_means, yerr=row_stds, fmt='-o', capsize=5)

# Customizing the plot
plt.title('PV-006710095262 Antagonism Curve of HCRTR2 Activation')
plt.xlabel('[PV-006710095262] (M)')
plt.ylabel('Luminescence (RLU)')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.grid(True)
plt.legend()

#set scale
plt.xscale('log')

#OxA average luminescence from experiment
positive_control = 137720
#DMSO average luminescence from experiment
negative_control = 38870

#add lines for OxA positive control and DMSO negative control
positive_line = plt.axhline(y=positive_control, color='green', linestyle='--', label='1e-7M OxA')
negative_line = plt.axhline(y=negative_control, color='red', linestyle='--', label='DMSO')

plt.legend(handles=[positive_line, negative_line], loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```
