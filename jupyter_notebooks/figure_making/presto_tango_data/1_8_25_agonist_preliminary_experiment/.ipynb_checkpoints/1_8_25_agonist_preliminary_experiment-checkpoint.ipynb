```python
#initial import of relevant packages for graphing
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import FuncFormatter
#https://stackoverflow.com/questions/32085531/importing-a-csv-file-into-python-to-make-scatterplots-and-histograms
```

```python
#read in the data
df = pd.read_csv("1_2_25_good_peptide_curve_organized_data_for_plot.csv", index_col=0)  # Set the first column as row index
```

```python
df
```

```python
#get mean and std dev
column_means = df.mean(axis=0)  # Mean of each column
column_stds = df.std(axis=0)    # Standard deviation of each column
```

```python
#make plot that include 1e-6 replicate 1 
#Well may have been compromised, as well volume looked greater than other wells, and point may be worth throwing out

plt.figure(figsize=(10, 6))
plt.errorbar(column_means.index, column_means, yerr=column_stds, fmt='-o', capsize=5)

# Customizing the plot
plt.title('HCRTR2 Activation Curve With OxA')
plt.xlabel('[OxA] (M)')
plt.ylabel('Luminescence (RLU)')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.grid(True)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
