```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# load TSV with no header
df = pd.read_csv("/home/ari.ginsparg-umw/ginsparg_thymelab_thesis/ordered_ligs_shapedb_vs_tak925/7sr8_64b_library_ligs_centered_vs_tak_scores.tsv", sep="\t", header=None)

# column 0 = ligand name
# column 1 = score
scores = df[1]

# make histogram
plt.figure()
plt.hist(scores, bins=20)
plt.xlabel("Shape similarity score")
plt.ylabel("Count")
plt.title("ShapeDB similarity distribution")
plt.show()
#/home/ari.ginsparg-umw/ginsparg_thymelab_thesis/ordered_ligs_shapedb_vs_tak925/agonists_only.tsv
```

```python
plt.figure(figsize=(6,4))
plt.hist(scores, bins=20, edgecolor="black")
plt.xlabel("Shape similarity score")
plt.ylabel("Number of ligands")
plt.title("Distribution of Shape Similarity Scores")
plt.tight_layout()
plt.show()
```

```python
tak_score = 0.62  # example

plt.hist(scores, bins=30, edgecolor="black")
plt.axvline(tak_score, color="red", linestyle="--", label="TAK-925")
plt.legend()
plt.show()
```
