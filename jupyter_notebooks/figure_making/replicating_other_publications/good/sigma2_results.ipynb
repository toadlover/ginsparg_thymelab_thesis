```python

dirname = "Structures_of_the_sigma2_receptor_enable_docking_for_bioactive_ligand_discovery"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# load file (no header)
df = pd.read_csv("../../../../benchmarking_other_papers/papers/" + dirname + "/analysis_space/best_ddg.csv", header=None, usecols=range(9))

# if there is an extra empty trailing column, drop it
df = df.dropna(axis=1, how="all")

# column 0 is the true class: 1 = published ligand, 0 = background
y_true = df[0].astype(int).to_numpy()

# since file is already sorted best -> worst, create a score from row order
n = len(df)
y_score = np.arange(n, 0, -1)

# ROC + AUC
fpr, tpr, thresholds = roc_curve(y_true, y_score)
roc_auc = auc(fpr, tpr)

print(f"Total compounds: {n}")
print(f"Published ligands: {y_true.sum()}")
print(f"Background ligands: {(y_true == 0).sum()}")
print(f"AUC: {roc_auc:.4f}")

# plot
plt.figure(figsize=(6,5))
plt.step(fpr, tpr, where="post")
plt.plot([0, 1], [0, 1], linestyle="--")

plt.xlabel("False Positive Rate", fontsize=14)
plt.ylabel("True Positive Rate", fontsize=14)
#plt.title("ROC Curve")

plt.tick_params(axis='both', labelsize=14)

# AUC box in bottom-right
plt.text(
    0.96, 0.04,
    f"AUC = {roc_auc:.4f}",
    transform=plt.gca().transAxes,
    ha="right",
    va="bottom",
    fontsize=20,
    bbox=dict(boxstyle="round", facecolor="white", edgecolor="black")
)

plt.tight_layout()
plt.show()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# load file (no header)
df = pd.read_csv("../../../../benchmarking_other_papers/papers/" + dirname + "/analysis_space/best_interactions.csv", header=None, usecols=range(9))

# if there is an extra empty trailing column, drop it
df = df.dropna(axis=1, how="all")

# column 0 is the true class: 1 = published ligand, 0 = background
y_true = df[0].astype(int).to_numpy()

# since file is already sorted best -> worst, create a score from row order
n = len(df)
y_score = np.arange(n, 0, -1)

# ROC + AUC
fpr, tpr, thresholds = roc_curve(y_true, y_score)
roc_auc = auc(fpr, tpr)

print(f"Total compounds: {n}")
print(f"Published ligands: {y_true.sum()}")
print(f"Background ligands: {(y_true == 0).sum()}")
print(f"AUC: {roc_auc:.4f}")

# plot
plt.figure(figsize=(6,5))
plt.step(fpr, tpr, where="post")
plt.plot([0, 1], [0, 1], linestyle="--")

plt.xlabel("False Positive Rate", fontsize=14)
plt.ylabel("True Positive Rate", fontsize=14)
#plt.title("ROC Curve")

plt.tick_params(axis='both', labelsize=14)

# AUC box in bottom-right
plt.text(
    0.96, 0.04,
    f"AUC = {roc_auc:.4f}",
    transform=plt.gca().transAxes,
    ha="right",
    va="bottom",
    fontsize=20,
    bbox=dict(boxstyle="round", facecolor="white", edgecolor="black")
)

plt.tight_layout()
plt.show()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# load file (no header)
df = pd.read_csv("../../../../benchmarking_other_papers/papers/" + dirname + "/analysis_space/best_real_count.csv", header=None, usecols=range(9))

# if there is an extra empty trailing column, drop it
df = df.dropna(axis=1, how="all")

# column 0 is the true class: 1 = published ligand, 0 = background
y_true = df[0].astype(int).to_numpy()

# since file is already sorted best -> worst, create a score from row order
n = len(df)
y_score = np.arange(n, 0, -1)

# ROC + AUC
fpr, tpr, thresholds = roc_curve(y_true, y_score)
roc_auc = auc(fpr, tpr)

print(f"Total compounds: {n}")
print(f"Published ligands: {y_true.sum()}")
print(f"Background ligands: {(y_true == 0).sum()}")
print(f"AUC: {roc_auc:.4f}")

# plot
plt.figure(figsize=(6,5))
plt.step(fpr, tpr, where="post")
plt.plot([0, 1], [0, 1], linestyle="--")

plt.xlabel("False Positive Rate", fontsize=14)
plt.ylabel("True Positive Rate", fontsize=14)
#plt.title("ROC Curve")

plt.tick_params(axis='both', labelsize=14)

# AUC box in bottom-right
plt.text(
    0.96, 0.04,
    f"AUC = {roc_auc:.4f}",
    transform=plt.gca().transAxes,
    ha="right",
    va="bottom",
    fontsize=20,
    bbox=dict(boxstyle="round", facecolor="white", edgecolor="black")
)

plt.tight_layout()
plt.show()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# load file (no header)
df = pd.read_csv("../../../../benchmarking_other_papers/papers/" + dirname + "/analysis_space/best_real_ratio.csv", header=None, usecols=range(9))

# if there is an extra empty trailing column, drop it
df = df.dropna(axis=1, how="all")

# column 0 is the true class: 1 = published ligand, 0 = background
y_true = df[0].astype(int).to_numpy()

# since file is already sorted best -> worst, create a score from row order
n = len(df)
y_score = np.arange(n, 0, -1)

# ROC + AUC
fpr, tpr, thresholds = roc_curve(y_true, y_score)
roc_auc = auc(fpr, tpr)

print(f"Total compounds: {n}")
print(f"Published ligands: {y_true.sum()}")
print(f"Background ligands: {(y_true == 0).sum()}")
print(f"AUC: {roc_auc:.4f}")

# plot
plt.figure(figsize=(6,5))
plt.step(fpr, tpr, where="post")
plt.plot([0, 1], [0, 1], linestyle="--")

plt.xlabel("False Positive Rate", fontsize=14)
plt.ylabel("True Positive Rate", fontsize=14)
#plt.title("ROC Curve")

plt.tick_params(axis='both', labelsize=14)

# AUC box in bottom-right
plt.text(
    0.96, 0.04,
    f"AUC = {roc_auc:.4f}",
    transform=plt.gca().transAxes,
    ha="right",
    va="bottom",
    fontsize=20,
    bbox=dict(boxstyle="round", facecolor="white", edgecolor="black")
)

plt.tight_layout()
plt.show()
```
