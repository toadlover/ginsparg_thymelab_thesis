```python
#initial import of relevant packages for graphing
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import FuncFormatter
#https://stackoverflow.com/questions/32085531/importing-a-csv-file-into-python-to-make-scatterplots-and-histograms
```

```python
#data from top 3M ligands from shapedb run on all 8 conformer shapes
#data transcribed from csv file
top_3m_data = [[1,5371667],[2,4052001],[3,444538],[4,1808134],[5,202510],[6,149330],[7,7093],[8,0]]

#convert to dataframe
df = pd.DataFrame(top_3m_data, columns=['X', 'Y'])
```

```python
#derive the total number of ligands
total_ligands = 0
for pair in top_3m_data:
    total_ligands = pair[1] + total_ligands
    
print("Total ligands", total_ligands)
```

```python
total_ligands_str = "12,035,273"
```

```python
# Create a bar plot
plt.bar(df['X'], df['Y'])
plt.xlabel('Ligands Found By Input Shapes')
plt.ylabel('Conformers')
plt.title('Conformers Appearance In Agonist Top 3,000,000 Shape Similarity')

# Set y-axis to display in scientific notation with individual tick values
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
plt.gca().yaxis.get_major_formatter().set_scientific(False)
plt.gca().yaxis.get_major_formatter().set_powerlimits((0,0))

# Set y-axis to display numbers with commas
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))

#add note for the text
#plt.text(3, 1, "Conformers: " + str(total_ligands), fontsize=12, color='black')
plt.annotate("Total Unique Conformers = " + str(total_ligands_str), xy=(0.245, -0.18), xycoords='axes fraction', fontsize=9, color='black')
plt.yscale('log')

plt.show()
```
