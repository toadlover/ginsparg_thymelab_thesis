```python
#imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap, TwoSlopeNorm, BoundaryNorm
import numpy as np
```

```python
#make custom color map
#custom_cmap = LinearSegmentedColormap.from_list("thresholded", ["magenta", "white", "green"])
#colors = ["magenta", "white", "green"]
#bounds = [-5, -1, 1, 5]  # Color boundaries
#norm = BoundaryNorm(boundaries=bounds, ncolors=256, clip=True)
#cmap = LinearSegmentedColormap.from_list("thresholded", colors, N=256)

colors = [
    (0.00, "magenta"),
    (0.40, "white"),  # point just before 0
    (0.60, "white"),  # point just after 0
    (1.00, "green")
]
cmap = LinearSegmentedColormap.from_list("magenta_white_green", colors)
norm = TwoSlopeNorm(vmin=-2, vcenter=0, vmax=2)
```

```python
#initial inport with numberofbouts
df = pd.read_csv("cleancol_ari_data_numberofbouts_4code.csv", index_col=0)
df = df[['day0night', 'day1day']]
df = df[~df.index.str.contains("clear", case=False)]
df = df.sort_index()
df = df * -1
```

```python
df
```

```python
plt.figure(figsize=(8, len(df) * 0.4))

#norm = TwoSlopeNorm(vmin=df.min().min(), vcenter=0, vmax=df.max().max())

ax = sns.heatmap(df, annot=False, cmap=cmap, norm=norm,   cbar_kws={'label': 'SSMD Score'})  # You can change colormap (e.g., 'coolwarm', 'RdBu', etc.)
plt.title("Number of Bouts (dpix)", fontsize=14)
plt.ylabel("Drugs", fontsize=14)
#plt.xlabel("Metrics")
plt.yticks(rotation=0, fontsize=14)
plt.gca().xaxis.set_ticks_position('top')     # Set ticks to top
plt.gca().xaxis.set_label_position('top')     # Set label to top

# Access the colorbar
cbar = ax.collections[0].colorbar

# Set font size for label and ticks
cbar.ax.tick_params(labelsize=12)                 # Tick values font size
cbar.set_label("SSMD", fontsize=14)         # Label font size

plt.tight_layout()
plt.show()
```
