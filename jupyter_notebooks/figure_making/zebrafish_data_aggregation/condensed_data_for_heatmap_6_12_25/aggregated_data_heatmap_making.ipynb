```python
#imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#initial inport with numberofbouts
df = pd.read_csv("ari_data_numberofbouts_.csv", index_col=0)
df = df[['night_avg', 'day_avg']]
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap="coolwarm")  # You can change colormap (e.g., 'coolwarm', 'RdBu', etc.)
plt.title("Number of Bouts (dpix)")
plt.ylabel("Drugs")
plt.xlabel("Metrics")
plt.tight_layout()
plt.show()
```
